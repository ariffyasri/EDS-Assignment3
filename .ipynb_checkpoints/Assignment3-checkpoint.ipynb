### Folium

[Folium](https://github.com/python-visualization/folium) makes it easy to visualize data that's been manipulated in Python on an interactive Leaflet map. It enables both the binding of data to a map for choropleth visualizations as well as passing Vincent/Vega visualizations as markers on the map.

```python
import pandas as pd
import re
import numpy as np
import folium
import matplotlib.pyplot as plt
%matplotlib inline
from bs4 import BeautifulSoup
import requests
```

```python
mapit = None
latlon = [ (3.144473, 101.708722), (3.144473, 101.708722),(3.135732, 101.686989)]
for coord in latlon:
    mapit = folium.Map( location=[ coord[0], coord[1] ] )
    folium.Marker([3.144473, 101.708722], icon=folium.Icon(color='blue',icon='star'), popup='Federal Hotel').add_to(mapit)
    folium.Marker([3.156374, 101.714579], icon=folium.Icon(color='green',icon='info-sign') , popup='Mandarin Oriental').add_to(mapit)
    folium.Marker([3.135732, 101.686989], icon=folium.Icon(color='red',icon='star') , popup='Le Meridien').add_to(mapit)
mapit
```

```python
# scrape from lelong

# find the pattern for the first page
url = 'https://www.lelong.com.my/catalog/all/list/?TheKeyword=laptop&CategoryID=764&D='

# write a loop to scrape from page 1 to the last page

#     Product Name/Product Title  - done
#     Amount/Price - done
#     Brand - otw
#     Comments/Reviews - otw
#     Number of views - done

def get_type(name):
    list_type = ['battery', 'adapter', 'keyboard']
    for types in list_type:
        types = str.lower(types)
        name = str.lower(name)
        regex = re.search("("+types+")", name)
        if regex:
            return str(regex.group())
    return 'laptop'

def get_brand(name):
    list_brand = ['ASUS','ACER','HP COMPAQ','DELL','Lenovo','HP','OEM','Asus','COMPAQ','Acer','LENOVO','TOSHIBA','Dell','Toshiba','MSI']
    for brand in list_brand:
        brand = str.lower(brand)
        name = str.lower(name)
        regex = re.search("("+brand+")", name)
        if regex:
            return str(regex.group())
    return 'no brand'

product_name = []
product_price = []
product_view = []
product_brand = []
product_type = []
for page in range(1,20):
    url_page = url+str(page)
    scrape = requests.get(url_page)
    soup = BeautifulSoup(scrape.content, 'lxml')
    link_a = soup.find_all('div',{'class':'item','class':'summary'})
    link_view = soup.find_all('span', {'class':'hit fontsize12'})
    length = len(link_a)
    for i in range(0,length):
        name = link_a[i].a.get('title')
        price = link_a[i].a.b.get('data-price')
        view = re.search(r'(\d+)', link_view[i].get_text())
        view = view.group(0)
        brand = get_brand(name)
        types = get_type(name)
        product_name.append(name)
        product_price.append(price)
        product_view.append(view)
        product_brand.append(brand)
        product_type.append(types)
# write to csv
# convert the list to a pandas dataframe

df = pd.DataFrame({'name':product_name, 'price':product_price, 'view':product_view, 'brand':product_brand, 'type':product_type})
df
df.to_csv('output.csv', index=False)

```
